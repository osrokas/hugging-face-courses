```python
from transformers import AutoTokenizer
```

```python
tokenizer = AutoTokenizer.from_pretrained("bert-base-cased")

encoded_input = tokenizer("Hello, I'm a single sentence!")
```

```python
encoded_input
```

```python
tokenizer.decode(encoded_input["input_ids"])
```

```python
encoded_input = tokenizer("How are you?", "I'm fine, thank you!", return_tensors="pt")
print(encoded_input)
```

## Because vectors are different lengths, we add padding to tokenizer.

```python
encoded_input = tokenizer(
    ["How are you?", "I'm fine, thank you!"], padding=True, return_tensors="pt"
)
print(encoded_input)
```

## If we have sentences longer than model context length, we use truncate param.

```python
encoded_input = tokenizer(
    "This is a very very very very very very very very very very very very very very very very very very very very very very very very very very very very very very very very very very very very very very very very very very very very very very very very very long sentence.",
    truncation=True,
)
print(encoded_input["input_ids"])
```

```python
encoded_input = tokenizer(
    ["How are you?", "I'm fine, thank you!"],
    padding=True,
    truncation=True,
    max_length=5,
    return_tensors="pt",
)
print(encoded_input)
```

## Multiple sentences tokenization

```python
from transformers import AutoModelForSequenceClassification
```

```python
checkpoint = "distilbert-base-uncased-finetuned-sst-2-english"
tokenizer = AutoTokenizer.from_pretrained(checkpoint)
model = AutoModelForSequenceClassification.from_pretrained(checkpoint)
```

```python
sequence = "I've been waiting for a HuggingFace course my whole life."
```

```python
tokens = tokenizer.tokenize(sequence)
```

```python
tokens
```

```python
ids = tokenizer.convert_tokens_to_ids(tokens)
```

```python
ids
```

```python
import torch
```

```python
input_ids = torch.tensor(ids)
```

```python
input_ids
```

```python
model(input_ids)
```

```python
tokenized_inputs = tokenizer(sequence, return_tensors="pt")
print(tokenized_inputs["input_ids"])
```

Because model can get multiple sentences for an input, it will throw error then we pass only one tokenized sentence.
We need to wrap a sentence in higher dimension

Putting sentences into array, called batching

```python

input_ids = torch.tensor([ids])
```

```python
model(input_ids)
```

```python
batched_ids = [ids, ids]
```

```python
model(torch.tensor(batched_ids))
```

### Tensor should also be same length

```python
model(torch.tensor(batched_ids))
```

```python
model(torch.tensor(batched_ids))
```

```python
sequence1_ids = [[200, 200, 200]]
sequence2_ids = [[200, 200]]
batched_ids = [
    [200, 200, 200],
    [200, 200, tokenizer.pad_token_id],
]
```

## Attention masks


When we have 1 - it means attention layer should take into account that token id
If 0 - ignore this token in attetion layer.

```python
batched_ids = [
    [200, 200, 200],
    [200, 200, tokenizer.pad_token_id],
]

attention_mask = [
    [1, 1, 1],
    [1, 1, 0],
]
```

```python
model(torch.tensor(batched_ids), attention_mask=torch.tensor(attention_mask))
```

Model can only apply predictions on sentences that are defines context size. 
Mostly 512 - 1024 tokens. 
To avoid error use proper context size model or truncate sentence,

```python
max_sequence_length = 512
sequence = sequence[:max_sequence_length]
```
